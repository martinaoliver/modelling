```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context('poster')
def dose_response(D, Kda):
    Va=100; nda=2
    A = 1+Va * (1/(1+(Kda/D)**nda))
    return A

D = np.linspace(1, 10, 100)
D_larger = np.logspace(-1, 2, 100)
Kda_list = np.logspace(-3, 3, 100)
for Kda in Kda_list:
    if Kda<np.amin(D) or Kda>np.amax(D):
        color='khaki'
    if Kda<np.amin(D)*0.1 or Kda>np.amax(D)*10:
        color='indianred'
        # print('aaa')
    if Kda>=np.amin(D) and Kda<=np.amax(D):
        color='mediumaquamarine'
    A = dose_response(D, Kda)
    plt.plot(D, A, label='Kda = {}'.format(Kda),color=color)
    plt.xscale('log')
plt.ylabel('[X*]')
plt.xlabel('[A]')
plt.savefig('balanced_vs_unbalanced_doseresponse.pdf')
plt.show()
```
