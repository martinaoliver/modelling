```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft

```

Step function

```python
x1 = np.linspace(-3,3,1200)
# y = [0,1,1,1,0]
# fx = np.sin(x)
y1=[]
for n in range(1200):
    if n<400:
        y1.append(0)
    elif n<800:
        y1.append(1)
    else:
        y1.append(0)
```

```python
x2 = np.linspace(-3,3,1200)
# y = [0,1,1,1,0]
# fx = np.sin(x)
y2=[]
for n in range(1200):
    if n<200:
        y2.append(0)
    elif n<1000:
        y2.append(1)
    else:
        y2.append(0)

```

```python
fig,ax = plt.subplots()
plt.step(x1,y1)
plt.show()
plt.step(x2,y2)
plt.show()

# ax[1]=plt.plot(x2,y2)
```

```python
fft(y1)
plt.hist(fft(y1),bins=120)
plt.show()

fft(y2)
plt.hist(fft(y2),bins=120)
plt.show()
fft(y1)

```

## Sinusoidal

```python
x = np.linspace(0,2*np.pi,100)
fx = np.sin(x)
plt.plot(x,fx)
```

```python
FT=fft(fx)
FT_real=np.real(FT)
FT_imag=np.imag(FT)

# len(FT_real)
# FT_real
# plt.hist(FT)
plt.scatter(x,FT_real)
plt.show()
plt.scatter(x,FT_imag)
```

```python
x = np.linspace(0,2*np.pi,100)
fx = np.cos(x)
plt.plot(x,fx)
```

```python
FT=fft(fx)
FT_real=np.real(FT)
FT_imag=np.imag(FT)

# len(FT_real)
# FT_real
# plt.hist(FT)
plt.scatter(x,FT_real)
plt.show()
plt.scatter(x,FT_imag)
```

### Case 1

```python
x=np.linspace(0,5,1000)
y=np.linspace(0,5,1000)
X,Y=np.meshgrid(x,y)

fxy= np.sin(4*np.pi*X)+np.cos(6*np.pi*Y)
plt.subplot(121)
plt.imshow(fxy)

#The inverse of fftshift. Although identical for even-length x, the functions differ by one sample for odd-length x.
ft = np.fft.ifftshift(fxy) 

#Compute the 2-dimensional discrete Fourier Transform (fast fourier transform).
ft = np.fft.fft2(ft)

#Shift the zero-frequency component to the center of the spectrum.
ft = np.fft.fftshift(ft)

plt.subplot(122)
plt.imshow(np.abs(ft))
plt.xlim(480,520)
plt.ylim(480,520)

plt.show()
```
