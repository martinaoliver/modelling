```python
import numpy as np
import pandas as pd

circuit=2
variant=0
n_parametersets = 1000
df = pd.read_pickle("lsa_df_circuit%r_variant%r_%rparametersets_wrongindex.pkl"%(circuit,variant,n_parametersets))
# df = df.reset_index()
tupled_index =  [tuple(l) for l in df.index]
```

```python
multi_index = pd.MultiIndex.from_tuples(tupled_index) 
df = df.set_index(multi_index)
df.to_pickle("lsa_df_circuit%r_variant%r_%rparametersets.pkl"%(circuit,variant,n_parametersets))
```

```python
df.loc[df['class'] == 'Turing I']
```

```python
def stability (row):
    if any(x > 0 for x in np.real(row['k0_eig'])):
        return 'Unstable'
    elif any(x == 0 for x in np.real(row['k0_eig'])):
        return 'Neutral'
    elif any(x < 0 for x in np.real(row['k0_eig'])):
        return 'Stable'

df['stability'] = df.apply (lambda row: stability(row), axis=1)

def system_type(row):
    ordered_k0_eig = row['k0_eig']
    ordered_k0_eig.sort()

    #Stable
    if row['stability'] == 'Stable':
        if np.imag(ordered_k0_eig[-1]) == 0:
            return 'Stable point'
        elif np.imag(ordered_k0_eig[-1]) != 0:
            return 'Stable spiral'
    
    #Neutral
    if row['stability'] == 'Neutral':
        if np.imag(ordered_k0_eig[-1]) == 0:
            return 'Neutral point'
        elif np.imag(ordered_k0_eig[-1]) != 0:
            return 'Neutral center'
    #Unstable
    if row['stability'] == 'Unstable':
        if np.imag(ordered_k0_eig[-1]) == 0:
            if np.real(ordered_k0_eig[-2]) < 0:
                return 'Saddle point'
            elif np.real(ordered_k0_eig[-2]) > 0:
                return 'Unstable point'
            else:
                return 'Nan'
        elif np.imag(ordered_k0_eig[-1]) != 0:
            return 'Unstable spiral'
    
    else:
        return 'NaN'
    
df['system type'] = df.apply (lambda row: system_type(row), axis=1)

  
df
```
