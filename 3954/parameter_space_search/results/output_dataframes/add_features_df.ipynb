```python
import numpy as np
import pandas as pd

```

```python
circuit=2
variant=0
n_parametersets = 10000
df = pd.read_pickle("lsa_df_circuit%r_variant%r_%rparametersets.pkl"%(circuit,variant,n_parametersets))
tupled_index =  [tuple(l) for l in df.index]
```

```python
multi_index = pd.MultiIndex.from_tuples(tupled_index) 
df = df.set_index(multi_index)
# df.to_pickle("lsa_df_circuit%r_variant%r_%rparametersets.pkl"%(circuit,variant,n_parametersets))
```

```python
df
```

```python
for parID in np.unique(df.index.get_level_values(0)):
        
    if 'Turing I' in df.loc[[parID]]['class'].values:
        df.loc[(parID),'turing_type'] = 'Turing I'
        
    elif 'Turing II' in df.loc[[parID]]['class'].values:
        df.loc[(parID),'turing_type'] = 'Turing II'
        
    elif 'Turing Oscillatory' in df.loc[[parID]]['class'].values:
        df.loc[(parID),'turing_type'] = 'Turing Oscillatory'
    

df
```

```python
df = df.drop(['system type', 'stability'], axis=1)
```

```python
df['class'].value_counts()
```

```python
def ss_stability (row):
    
    k0_eig = row['k0_eig']

    if np.any(np.isnan(k0_eig)):
        return 'no steady state'
    elif any(x > 0 for x in np.real(k0_eig)):
        return 'Unstable'
    elif any(x == 0 for x in np.real(k0_eig)):
        return 'Neutral'
    elif any(x < 0 for x in np.real(k0_eig)):
        return 'Stable'

df['ss_stability'] = df.apply (lambda row: ss_stability(row), axis=1)

def ss_type(row):
    ordered_k0_eig = row['k0_eig']
    ordered_k0_eig.sort()

    #Stable
    if row['ss_stability'] == 'Stable':
        if np.imag(ordered_k0_eig[-1]) == 0:
            return 'Stable point'
        elif np.imag(ordered_k0_eig[-1]) != 0:
            return 'Stable spiral'
    
    #Neutral
    if row['ss_stability'] == 'Neutral':
        if np.imag(ordered_k0_eig[-1]) == 0:
            return 'Neutral point'
        elif np.imag(ordered_k0_eig[-1]) != 0:
            return 'Neutral center'
    #Unstable
    if row['ss_stability'] == 'Unstable':
        if np.imag(ordered_k0_eig[-1]) == 0:
            if np.real(ordered_k0_eig[-2]) < 0:
                return 'Saddle point'
            elif np.real(ordered_k0_eig[-2]) > 0:
                return 'Unstable point'
            else:
                return 'Nan'
        elif np.imag(ordered_k0_eig[-1]) != 0:
            return 'Unstable spiral'
    
    else:
        return 'NaN'
    
df['ss_type'] = df.apply (lambda row: ss_type(row), axis=1)

  
df
```
