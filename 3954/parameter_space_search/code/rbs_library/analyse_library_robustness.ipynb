### Distributions of probability for each case

```python
states = ['turing I', 'turing I hopf', 'turing I oscillatory'] 
turing_ratio=[]
original_index_list = df['original_index'].unique()
for n in original_index_list[:-1000]:
    original_index_df = df.loc[df['original_index'] == n]
    turing_original_index = original_index_df.loc[original_index_df['system_class'].isin(states)]
    turing_ratio.append((len(turing_original_index)/len(original_index_df)))
```

```python
sns.histplot(turing_ratio, bins=200)
plt.yscale('log')
plt.ylabel('Count (log)')
plt.xlabel('Turing vs Non-Turing ratio')
plt.savefig('rbs_turing_ratio_histogram', dpi=1000)
```
