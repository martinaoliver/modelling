### which original indexes lead to turing with library?

```python
states = ['turing I', 'turing I hopf', 'turing I oscillatory'] 
turing_count=0
for n in range(1,101):
    print(n)
    original_index_df = df.loc[df['original_index'] == n]
    A = original_index_df['system_class'].value_counts()
    print(A)
    print(any(x in A for x in states))
    if any(x in A for x in states):
        turing_count +=1
#     if any(original_index_df['system_class'] == "simple stable" or original_index_df['system_class']==):
#         print('a`sfzgfhjf')
    print()
    print()
print(turing_count)
```
