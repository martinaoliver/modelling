```python
import numpy as np
from scipy.sparse import spdiags, eye
from scipy import signal
import matplotlib.pyplot as plt
import scipy
import timeit
from scipy.linalg import solve, lu_factor, lu_solve


```

```python
D=1; 

#space
Lx=1; #Diffusion constant; Domain length
J = 10
dx = Lx/(J-1)


#time
T = 10
N = 1000
dt = float(T/(N-1))




# x=-Lx + (0:Nx)*dx; #x values on the grid
# dt=(dx)**2/(2*D); #borderline stability of FTCS scheme 
alpha=dt*D/dx**2;

nx=J
dt
```

```python
data = np.array([-alpha*np.ones(nx), 2*(1+alpha)*np.ones(nx), -alpha*np.ones(nx)])
diags = np.array([-1, 0, 1])
I = eye(nx).toarray()
print(I)
A = spdiags(data, diags,nx,nx).toarray()
A[[0,nx-1],:] = I[[0,nx-1],:]
print( A)
```

```python
sigma=Lx/1; #stedv of the gaussian - sharp gaussian as initial condition
u = signal.gaussian(nx,sigma)*100
u=u.transpose()
plt.plot(np.linspace(0,nx,nx),u)# xlabel('$x$','Interpreter','latex','FontSize',14);
plt.title('initial condition')
plt.xlim(0,10)
u1=u
u


```

```python
from scipy.linalg import lu_factor, lu_solve
print(A)
# luA, piv = lu_factor(A)
start = timeit.timeit()
for m in range(N):
    plt.plot(np.linspace(0,nx,nx),u,label=m)
    b = np.array([0]+[[alpha*u[n-1] + 2*(1-alpha)*u[n] + alpha*u[n+1]][0] for n in range (1,J-1)]+[0])
    u = solve(A, b)
#     u = scipy.linalg.lu_solve((luA,piv),b)
end = timeit.timeit()
# plt.legend()
np.linspace(0,nx,nx)
plt.xlim(0,10)
print(end - start)
```
